```python
import numpy as np
import copy
import random

goal = np.array([[1, 2, 3], [4, 0, 5], [6, 7, 8]])
ROWS = 3
COLS = 3

def print_puzzle(puzzle):
  for i in range(ROWS):
    for j in range(COLS):
      print(puzzle[i][j], end = " ")
    print()

def puzzle_generator():
    # Use any starting combination..
    random_seed = random.randrange(0, 10)
    puzzles = ([
              [[1, 0, 3], [4, 2, 5], [6, 7, 8]],
              [[1, 5, 2], [0, 4, 3], [6, 7, 8]],
              [[1, 5, 2], [4, 7, 3], [6, 8, 0]],
              [[1, 5, 0], [4, 3, 2], [6, 7, 8]],
              [[1, 5, 2], [4, 3, 8], [6, 7, 0]],
              [[4, 1, 2], [0, 5, 3], [6, 7, 8]],
              [[1, 5, 2], [0, 7, 3], [4, 6, 8]],
              [[1, 5, 2], [4, 7, 0], [6, 8, 3]],
              [[1, 5, 2], [4, 7, 0], [6, 8, 3]],
              [[4, 1, 2], [5, 0, 3], [6, 7, 8]]
            ])
    
    return puzzles[random_seed]

def is_goal(puzzle):
    for i in range(0, ROWS):
        for j in range(0, COLS):
            if puzzle[i][j] != goal[i][j]:
                return False

    print("\n*W O H O O !  Puzzle Matched!")
    print_puzzle(puzzle)
    return True


def is_already_visited(current_puzzle, visited):
    
    if len(visited) == 0:
        return False
    for i in range(len(visited)):
        counter = 0
        for j in range(0, ROWS):
            for k in range(0, COLS):
                if current_puzzle[j][k] == visited[i][j][k]:
                    counter += 1
        if counter == ROWS * COLS:
            print("\n* [ Already Visited! ] ")
            return True
    return False


def swap(puzzle, src_row, src_col, dest_row, dest_col):
    temp = copy.deepcopy(puzzle)
    temp[src_row][src_col], temp[dest_row][dest_col] = temp[dest_row][dest_col], temp[src_row][src_col]
    return temp

def DFS(puzzle):

    print("Starting puzzle..")
    print_puzzle(puzzle)
    
    stack = []
    visited = [] # To keep track of visited combinations
               
    count = 0          
    while not is_goal(puzzle):

      print("------------------------------------------------------------")
      if (count == 362880):
        print("NO SOLUTION FOUND! (After 9! = 362880 Iterations.)")
        break

      # Search ZERO-0
      row = col = -1
      for i in range(0, ROWS):
          for j in range(0, COLS):
            if puzzle[i][j] == 0:
              row = i
              col = j
              break

      # Move Right
      if col+1 < COLS and not is_already_visited(swap(puzzle, row, col, row, col + 1), visitedArray):
        stack.append(swap(puzzle, row, col, row, col+1))
        print("Pushing right!", end=" ")
      # Move Left
      if col > 0 and not is_already_visited(swap(puzzle, row, col, row, col - 1), visitedArray):
        stack.append(swap(puzzle, row, col, row, col-1))
        print("Pushing left!", end=" ")
      # Move Down
      if row+1 < ROWS and not is_already_visited(swap(puzzle, row, col, row + 1, col), visitedArray):
        stack.append(swap(puzzle, row, col, row + 1, col))
        print("Pushing down!", end=" ")  
      # Move Up
      if row > 0 and not is_already_visited(swap(puzzle, row, col, row - 1, col), visitedArray):
        stack.append(swap(puzzle, row, col, row - 1, col))
        print("Pushing up!", end=" ")

      visited.append(puzzle)
      puzzle = stack.pop(0)

      count += 1
      print("\n<Iteration>: ", count)

random_puzzle = puzzle_generator()
DFS(random_puzzle)
```
